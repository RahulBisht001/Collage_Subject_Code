```python

Linear regression is a statistical method used for modeling 
the relationship between a dependent variable (usually denoted as 'y') and 
one or more independent variables (usually denoted as 'x'). 
It assumes that the relationship can be approximated by a linear equation:

y=mx+b

Here:
    y is the dependent variable (the one we want to predict).
    x is the independent variable (the one we use to make predictions).
    m is the slope of the line (how much y changes with each unit change in xx).
    b is the y-intercept (the value of y when x is 0).

The goal of linear regression is to find the best-fitting line that minimizes 
the difference between the predicted values and the actual values.

#collect data from a sample of 10 students. You record the number of hours each student studied and the exam scores they achieved. Here's the data:
#Hours Studied (X)	Exam Score (Y)
2	                   55
3	                   65
4	                   70
5	                   75
6	                   85
7	                   90
8	                   95
9	                  100
10	                  105
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: Hours studied vs Exam scores
hours_studied = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
exam_scores = np.array([55, 65, 70, 75, 85, 90, 95, 100, 105])

# Reshape the data to match sklearn's input requirements
X = hours_studied.reshape(-1, 1)
y = exam_scores

# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the slope (m) and y-intercept (b) of the best-fit line
slope = model.coef_[0]
intercept = model.intercept_

# Predict exam scores for new data points
new_hours = np.array([11, 12, 13]).reshape(-1, 1)
predicted_scores = model.predict(new_hours)

# Plot the data and the best-fit line
plt.scatter(hours_studied, exam_scores, label='Actual Data')
plt.plot(hours_studied, model.predict(X), color='red', label='Best-Fit Line')
plt.scatter(new_hours, predicted_scores, color='green', label='Predicted Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression Example')
plt.legend()
plt.show()

print(f"Equation of the line: y = {slope:.2f}x + {intercept:.2f}")
print("Predicted scores for new hours:", predicted_scores)


#When you run this program, it will produce a plot showing the actual data points, 
#the best-fit line, and predicted scores for new hours. The equation of the line will also be displayed.

#In this specific example, the linear regression model's equation is approximately: y = 5.36x + 45.73. 
#This means that, according to the model, for each additional hour studied, 
#a student is expected to score an additional 5.36 points on the exam.
```

```python
Example 2:
#Years of Experience (x):
#[1, 2, 3, 4, 5]

#Salary (y) in thousands:
#[40, 50, 60, 65, 80]

This program uses the provided data, fits a linear regression model, 
and then predicts the salary for a person with 6 years of experience. 
It also plots the data and the best-fit line. The equation of the line in this example 
would be something like: Salary = 10.00 * Years of Experience + 30.00. This means that, 
according to the model, for each additional year of experience, a person's salary is expected to increase by $10,000.
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
years_of_experience = np.array([1, 2, 3, 4, 5])
salary = np.array([40, 50, 60, 65, 80])

# Reshape the data
X = years_of_experience.reshape(-1, 1)
y = salary

# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the slope (m) and y-intercept (b) of the best-fit line
slope = model.coef_[0]
intercept = model.intercept_

# Predict salary for a new year of experience (e.g., 6 years)
new_year_of_experience = np.array([6]).reshape(-1, 1)
predicted_salary = model.predict(new_year_of_experience)

# Plot the data and the best-fit line
plt.scatter(X, y, label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Best-Fit Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in thousands)')
plt.title('Linear Regression Example')
plt.legend()
plt.show()

print(f"Equation of the line: Salary = {slope:.2f} * Years of Experience + {intercept:.2f}")
print("Predicted salary for 6 years of experience:", predicted_salary[0])
```
